### Analyze "All Names" data that has been parsed into `Top-1000-Names.xlsx`

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
```

```python
males = pd.read_excel('Top-1000-Names.xlsx', sheet_name='Males', index_col=0)
males.index = males.index + 1  # index=rank

females = pd.read_excel('Top-1000-Names.xlsx', sheet_name='Females', index_col=0)
females.index = females.index + 1 # index=rank
```

```python
# Number of unique names in the top 1000, males and females
unique_males = set(males.to_numpy().flatten())
unique_males = sorted(unique_males)
unique_females = set(females.to_numpy().flatten())
unique_females = sorted(unique_females)

print(f'Number of Unique Names in the top 1000, Male: {len(unique_males)}')
print(f'Number of Unique Names in the top 1000, Female: {len(unique_females)}')
```
